```python
from PIL import Image

# Buka gambar yang ingin diubah menjadi grayscale
input_image_path = "input.jpg"
output_image_path = "output.jpg"
image = Image.open(input_image_path)

# Konversi gambar ke skala abu-abu
grayscale_image = image.convert("L")

# Simpan gambar skala abu-abu
grayscale_image.save(output_image_path)

# Tampilkan gambar skala abu-abu (opsional)
grayscale_image.show()
```

Resizing


```python
from PIL import Image

# Load the image
input_image_path = "input.jpg"
image = Image.open(input_image_path)

# Dapatkan lebar dan tinggi gambar dalam piksel
width, height = image.size

# Cetak lebar dan tinggi gambar
print(f"Lebar gambar: {width} piksel")
print(f"Tinggi gambar: {height} piksel")
```

```python
# Define the new dimensions (width x height) in pixels
new_width = 400
new_height = 200

# Resize the image
resized_image = image.resize((new_width, new_height))

# Save the resized image
output_image_path = "resized_output.jpg"
resized_image.save(output_image_path)
```

```python
from PIL import Image
from IPython.display import display

# Load the grayscale image
resize_image = Image.open("resized_output.jpg")

# Display the grayscale image
display(resize_image)

# Dapatkan lebar dan tinggi gambar dalam piksel
widthresize, heightresize = resize_image.size

# Cetak lebar dan tinggi gambar
print(f"Lebar gambar: {widthresize} piksel")
print(f"Tinggi gambar: {heightresize} piksel")
```

Blurring Image


```python
import cv2
import numpy as np
# Load the image
input_image_path = "input.jpg"
image = cv2.imread(input_image_path)

# Apply Gaussian blur
kernel_size = (5, 5)  # Adjust the kernel size as needed
blurred_image = cv2.GaussianBlur(image, kernel_size, 0)

# Save the blurred image
output_image_path = "blurred_output.jpg"
cv2.imwrite(output_image_path, blurred_image)

# Display the original and blurred images (optional)
from google.colab.patches import cv2_imshow
cv2_imshow(image)
cv2_imshow( blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Reducing Noise dengan LPF


```python
import cv2
import numpy as np

# Baca citra dengan noise (contoh: citra berumur)
image_with_noise = cv2.imread('noise.jpeg')  # Ganti 'citra_noise.jpg' dengan nama file citra Anda

# Terapkan filter Gaussian untuk mengurangi noise
kernal_size = (7,7)  # Ukuran kernel (biasanya ganjil)
sigma_x = 0  # Nilai sigma dalam arah X (0 untuk otomatis menghitung berdasarkan ukuran kernel)
sigma_y = 0  # Nilai sigma dalam arah Y (0 untuk otomatis menghitung berdasarkan ukuran kernel)

# Terapkan filter rata-rata (LPF) untuk mengurangi noise
image_without_noise = cv2.GaussianBlur(image_with_noise, kernel_size, sigma_x, sigma_y)

# Tampilkan citra asli dan citra tanpa noise
from google.colab.patches import cv2_imshow
cv2_imshow(image_with_noise)
cv2_imshow(image_without_noise)

# Tunggu sampai pengguna menekan tombol apapun
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Baca gambar dari Google Drive (ganti dengan path yang sesuai)
image_path = 'noise.jpeg'  # Ganti dengan path gambar Anda
image = cv2.imread(image_path)

# Konversi gambar ke grayscale (opsional, tergantung pada jenis gambar Anda)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Terapkan Median Filter
kernel_size = 3  # Cobalah berbagai ukuran kernel
image_filtered = cv2.medianBlur(gray, kernel_size)

# Tampilkan gambar asli dan hasil Median Filter
cv2_imshow(gray)
cv2_imshow(image_filtered)

# Simpan gambar hasil Median Filter (jika diperlukan)
cv2.imwrite('hasil_filtered.jpg', image_filtered)  # Ganti dengan path penyimpanan yang sesuai
```

Cropping Objek

```python
import cv2
from google.colab import files

# Upload gambar dari komputer Anda ke Google Colab
uploaded = files.upload()

# Tentukan nama file gambar yang diunggah
image_filename = list(uploaded.keys())[0]

# Muat gambar menggunakan OpenCV
image = cv2.imread(image_filename)
```

```python
x1, y1 = 100, 100  # Koordinat kiri atas
x2, y2 = 300, 300  # Koordinat kanan bawah
```

```python
cropped_image = image[y1:y2, x1:x2]

from matplotlib import pyplot as plt

# Konversi warna gambar dari BGR ke RGB (jika diperlukan)
cropped_image_rgb = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)

# Tampilkan gambar
plt.imshow(cropped_image_rgb)
plt.axis('off')
plt.show()
```
